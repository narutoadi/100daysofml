```python
# source - https://www.youtube.com/watch?v=szXbuO3bVRk&list=PLRqwX-V7Uu6bCN8LKrcMa6zF4FPtXyXYj&index=5

%matplotlib inline
import random
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
fig = plt.figure()
ax = plt.axes()
# generating 100 (x,y) values
siz = 100 # number of points
x = range(siz)
x = [xp/siz for xp in x]
y=[]
sets = 10
cardinality = int (siz/sets)
for i in range(0,sets):
    j=cardinality*(i)
    k=j+cardinality
    temp = [xp+j/100 for xp in x[j:k]]
    y.extend(temp)

#random.shuffle(y)
# oh my GOD!!!!!!!! This shuffling took my 2 hours, i was missing it earlier 
combined = list(zip(x, y))
random.shuffle(combined)

x[:], y[:] = zip(*combined)
print(len(x), len(y))
plt.scatter(x, y)
plt.show()
print(x)
print(y)
```

```python
# lets assume a slope and a constant
m=0
b=0
# learning rate defines how slow or fast to move
lr = .2
# error tells the direction and magnitude of tweak
print(m,b,lr,siz)
for i in range(0,siz):
    guess = m*x[i]+b
    error = y[i]-guess
    m=m+(error*x[i])*lr
    b=b+error*lr
    print("x=",x[i],"y=",y[i],"guess = ",guess)
    print(error)
    
# after loop we have got m and b, lets plot everything now
print(m,b)

    
```

```python
yls = [xp*m+b for xp in x]
plt.scatter(x,y,color='b')
plt.scatter(x,yls,color='r')
plt.show()
```
