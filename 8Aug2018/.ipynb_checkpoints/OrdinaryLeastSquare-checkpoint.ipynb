```python
# source - https://www.youtube.com/watch?v=szXbuO3bVRk&list=PLRqwX-V7Uu6bCN8LKrcMa6zF4FPtXyXYj&index=5

%matplotlib inline
import matplotlib.pyplot as plt
import random
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
# generating 100 (x,y) values
siz = 100 # number of points
x = np.random.randint(low=1, high=1000, size=siz).tolist()
y=[]
sets = 10
cardinality = int (siz/sets)
for i in range(0,sets):
    j=cardinality*(i)
    k=j+cardinality
    temp = [xp*(j+1) for xp in x[j:k]]
    y.extend(temp)

x=[xp/1000 for xp in x]
y=[yp/100000 for yp in y]
combined = list(zip(x, y))
random.shuffle(combined)
x[:], y[:] = zip(*combined)
print(len(x), len(y))
#random.shuffle(y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# lets calculate the slope m and constant b for line y=mx+b with least square formula
numerator = 0
denominator = 0
xbar = sum(x)/len(x)
ybar = sum(y)/len(y)
for i in range(0,siz):
    numerator += (x[i]-xbar)*(y[i]-ybar)
    denominator+= (x[i]-xbar)*(x[i]-xbar)

m= numerator/denominator
b= ybar-m
print(m,b)    
```

```python
# lets plot the line and the points on the same graph
# yls = y generated by least square method
yls = [xp*m+b for xp in x]
```

```python
plt.scatter(x,y,color='b')
plt.scatter(x,yls,color='r')
plt.show()
```
